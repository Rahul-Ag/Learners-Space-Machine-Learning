# Numpy Implementation

```python
import numpy as np

# Create a 2D Numpy array of size 1x3 with elements of your choice
arr1 = np.array([[1, 2, 3]])

# Create a Numpy array of length 50 with zeroes as its elements
arr2 = np.zeros(50)

# Create a Numpy array of size 3x2 with elements of your choice
arr3 = np.array([[4, 5],
                 [6, 7],
                 [8, 9]])

# Multiply arr1 and arr3 using Numpy functions
arr4 = np.matmul(arr1, arr3)

# Change the 5th element of arr2 to a different number
arr2[4] = 10

# Check conditions
if np.shape(arr4) == (1, 2) and arr2[4] != 0:
    print("Passed")
else:
    print("Fail")
```

# Pandas Implementation

```python
import pandas as pd

# Create a DataFrame from a dictionary
data = {
    'Name': ['Ramesh', 'Mahesh', 'Suresh'],
    'Age': [25, 30, 35],
    'City': ['Bangalore', 'Mumbai', 'Delhi']
}
df = pd.DataFrame(data)

# Display the first 2 rows of the data frame
print("First 2 rows of the DataFrame:")
print(df.head(2))  # head(2) gives the first 2 rows

# Print the 'Age' column
print("\nAge column:")
print(df['Age'])

# Filter rows where age is greater than 26
print("\nRows where age is greater than 26:")
print(df[df['Age'] > 26])

# Add a new column 'Country' with the value 'India' for all rows
df['Country'] = 'India'

print("\nDataFrame with 'Country' column added:")
print(df)

# Handling missing values example:

# Data with missing values
data1 = {
    'Name': ['Ramesh', 'Mahesh', 'Suresh'],
    'Age': [25, None, 35],  # introducing a missing value
    'City': ['Bangalore', 'Mumbai', 'Delhi']
}
df2 = pd.DataFrame(data1)

# Fill missing values in the 'Age' column with the mean age
mean_age = df2['Age'].mean()
df2['Age'].fillna(mean_age, inplace=True)

print("\nDataFrame with missing values filled with mean age:")
print(df2)
```

# Matplotlib Implementation

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 3, 4])
ypoints = np.array([2, 4, 6, 8])

# Plot these points without drawing a line
plt.figure(figsize=(8, 6))  # Optional: Adjust figure size
plt.plot(xpoints, ypoints, 'o')  # 'o' for markers without a line
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Plot of Points without Line')
plt.show()

# Plotting with star marker
plt.figure(figsize=(8, 6))  # Optional: Adjust figure size
plt.plot(xpoints, ypoints, marker='*', linestyle='None')  # '*' for star marker
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Plot of Points with Star Marker')
plt.show()

# Using fmt format: circular marker, red color, and dashed line
plt.figure(figsize=(8, 6))  # Optional: Adjust figure size
plt.plot(xpoints, ypoints, 'ro--')  # 'ro--' for red circles with dashed lines
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Plot of Points with Red Circles and Dashed Line')
plt.show()

# Create a scatter plot for xpoints and ypoints
plt.figure(figsize=(8, 6))  # Optional: Adjust figure size
plt.scatter(xpoints, ypoints)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Scatter Plot of Points')
plt.show()

# Set colors for scatter plot points: Blue, Green, Red, Yellow
colors = ['blue', 'green', 'red', 'yellow']
plt.figure(figsize=(8, 6))  # Optional: Adjust figure size
for i in range(len(xpoints)):
    plt.scatter(xpoints[i], ypoints[i], color=colors[i])
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Scatter Plot with Colored Points')
plt.show()
```

# Miscellaneous Modules Implementation

```python
import random
import numpy as np

# Set the seed of random to 20
random.seed(20)

arr1 = np.array([1, 24, 31, 45, 73, 81, 94, 25])

# Using the random module, pick 4 different random numbers from arr1 and return their sum
random_numbers = random.sample(list(arr1), 4)
sum_of_randoms = sum(random_numbers)

print("Randomly picked numbers:", random_numbers)
print("Sum of randomly picked numbers:", sum_of_randoms)
```
